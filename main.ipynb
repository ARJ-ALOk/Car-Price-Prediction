```python
#creating our data unique which is called encoding

for col in car_data.columns :
    print("Unique colums "+col)
    print(car_data[col])
    print("*******************************************")
```

```python
#creating a function to shorten the car's name to make encoding easier
def model_name(carName):
    carName = carName.split()[0]
    return carName

print(model_name(" Maruti Swift Dzire VDI"))
```

```python
#to apply the function on dataframe
car_data["name"] = car_data["name"].apply(model_name)
```

```python
car_data["name"].head()
```

```python
# del car_data["torque"]
car_data.drop(columns = ["torque"], inplace = True)
```

```python
#chenging from alphanumeric to numeric
def clean_data(value) : 
    value = value.split(' ')[0]
    value = value.strip()
    if value == '':
        value = 0 
    return float(value)
```

```python
car_data["mileage"] = car_data["mileage"].apply(clean_data)
car_data["engine"] = car_data["engine"].apply(clean_data)
car_data["max_power"] = car_data["max_power"].apply(clean_data)
```

```python
car_data.sample(5)
```

```python
le = LabelEncoder()
car_data["fuel"] = le.fit_transform(car_data["fuel"])
car_data["seller_type"] = le.fit_transform(car_data["seller_type"])
car_data["name"] = le.fit_transform(car_data["name"])
car_data["owner"] = le.fit_transform(car_data["owner"])
car_data["transmission"] = le.fit_transform(car_data["transmission"])
```

```python
car_data.sample(5)
```

```python
model = LinearRegression()

input_data = car_data.drop(columns=['selling_price'])
output_data = car_data['selling_price']

```

```python
X_train, X_test, y_train, y_test = train_test_split(
    input_data,
    output_data,
    test_size= 0.2
)
```

```python
model.fit(X_train,y_train)
```

```python
predict = model.predict(X_test)
print(predict)
```
